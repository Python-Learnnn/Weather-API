```python
# 8. Perform convolution, pooling and flattening operation on an input image and display both input, output images.
import cv2
import numpy as np
import matplotlib.pyplot as plt  # Import matplotlib for plotting
from skimage.measure import block_reduce  # Required for pooling operation
from sklearn.preprocessing import StandardScaler  # Required for flattening operation

# Step 1: Upload your input image to Google Colab
# Make sure to upload an image file named 'input.jpeg'

# Step 2: Read the input image
input_image = cv2.imread('pattern.jpg', cv2.IMREAD_GRAYSCALE)

# Step 3: Define the convolution kernel/filter
kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

# Step 4: Perform convolution operation
convolution_output = cv2.filter2D(input_image, -1, kernel)

# Step 5: Perform pooling operation
pooling_output = block_reduce(input_image, (2, 2), np.max)

# Step 6: Flatten the pooled output
flattened_output = StandardScaler().fit_transform(pooling_output.reshape(-1, 1)).reshape(pooling_output.shape)

# Step 7: Display input and output images in two rows with axes
plt.figure(figsize=(15, 10))  # Set the figure size

# First row: original and convolutional images
plt.subplot(2, 2, 1)
plt.imshow(input_image, cmap='gray')
plt.title('Original Image')
plt.axis('on')

plt.subplot(2, 2, 2)
plt.imshow(convolution_output, cmap='gray')
plt.title('Convolution Output')
plt.axis('on')

# Second row: pooling and flattened images with updated titles
plt.subplot(2, 2, 3)
plt.imshow(pooling_output, cmap='gray')
plt.title('Pooling Output')
plt.axis('on')

plt.subplot(2, 2, 4)
plt.imshow(flattened_output, cmap='gray')
plt.title('Flattened Output (After Pooling)')
plt.axis('on')

plt.tight_layout()  # Adjust the layout for better spacing
plt.show()  # Show the plot
```

```python
# 9. Perform max pooling operation on an input image and display both input, output image with size of image.
import cv2
import numpy as np
import matplotlib.pyplot as plt  # Import matplotlib for plotting
from skimage.measure import block_reduce  # Required for pooling operation

# Step 1: Read the input image
input_image = cv2.imread('pattern.jpg', cv2.IMREAD_GRAYSCALE)

# Step 2: Define the pooling size
pool_size = (2, 2)

# Step 3: Perform max pooling operation
pooling_output = block_reduce(input_image, pool_size, np.max)

# Step 4: Display input and output images with sizes
plt.figure(figsize=(12, 6))  # Set the figure size

# Plot original image
plt.subplot(1, 2, 1)
plt.imshow(input_image, cmap='gray')
plt.title(f'Original Image (Size: {input_image.shape[1]} x {input_image.shape[0]})')
plt.axis('on')

# Plot pooling output
plt.subplot(1, 2, 2)
plt.imshow(pooling_output, cmap='gray')
plt.title(f'Max Pooling Output (Size: {pooling_output.shape[1]} x {pooling_output.shape[0]})')
plt.axis('on')

plt.tight_layout()  # Adjust the layout for better spacing
plt.show()  # Show the plot
```

```python
# 10 . Perform average pooling operation on an input image and display both input, output image with size of image.
import cv2
import numpy as np
import matplotlib.pyplot as plt  # Import matplotlib for plotting
from skimage.measure import block_reduce  # Required for pooling operation

# Step 1: Read the input image
input_image = cv2.imread('pattern.jpg', cv2.IMREAD_GRAYSCALE)

# Step 2: Define the pooling size
pool_size = (2, 2)

# Step 3: Perform average pooling operation
pooling_output = block_reduce(input_image, pool_size, np.mean)

# Step 4: Display input and output images with sizes
plt.figure(figsize=(12, 6))  # Set the figure size

# Plot original image
plt.subplot(1, 2, 1)
plt.imshow(input_image, cmap='gray')
plt.title(f'Original Image (Size: {input_image.shape[1]} x {input_image.shape[0]})')
plt.axis('on')

# Plot pooling output
plt.subplot(1, 2, 2)
plt.imshow(pooling_output, cmap='gray')
plt.title(f'Average Pooling Output (Size: {pooling_output.shape[1]} x {pooling_output.shape[0]})')
plt.axis('on')

plt.tight_layout()  # Adjust the layout for better spacing
plt.show()  # Show the plot
```
